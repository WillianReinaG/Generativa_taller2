# RAG EcoMarket con LlamaIndex

Este notebook usa la misma base documental de `kb/` para mostrar el flujo RAG con LlamaIndex.

```python
from pathlib import Path
from dotenv import load_dotenv
from llama_index.core import SimpleDirectoryReader, VectorStoreIndex

ROOT = Path.cwd().resolve().parents[0]
load_dotenv(ROOT / '.env')
kb_dir = ROOT / 'kb'
kb_dir
```

```python
documents = SimpleDirectoryReader(input_dir=str(kb_dir)).load_data()
index = VectorStoreIndex.from_documents(documents)
query_engine = index.as_query_engine(similarity_top_k=4)
response = query_engine.query('Que dice la politica de devolucion para productos de higiene?')
print(response)
```
